Find the mean ,mode and median of a list

```python
# Question 5
list=[5,1,2,2,4,2,3,6]
n=len(list)
if(len(list)==0):
  print("Empty list")
else:

     """ Sorting """
     for i in range(n):
       for j in range(0,n-i-1):
         if(list[j]>list[j+1]):
            temp=list[j+1]
            list[j+1]=list[j]
            list[j]=temp
     print("Sorted list",list)

     #MEAN
     sum=0
     for a in list:
      sum+=a
     print("Mean value is:",float(sum/n))

     #MEDIAN
     if(n%2==0):
      med=(list[int((n/2)-1)]+list[int(n/2)])/2
     else:
      med=list[int(n/2)]
     print("Median of list is:",med)

     #MODE
     max=0
     ele=0
     for a in list:
      count=0
      for b in list:
        if(a==b):
          count+=1
        if(count>max):
          max=count
          ele=a
     print("Mode of list is:",ele)
```

Standard deviation

```python
# Question 6
a=[5,6,4,3]
mean=0
for i in a:
  mean+=i
mean=mean/len(a)
sq=0
for j in a:
  sq+=(j-mean)**2
sq=sq/len(a)
print("Standard deviation of list is:",sq**0.5)
```

Addition of two matrices

```python
# Question 7
matrix1=[]
row1=int(input("Enter the value of row of matrix1:"))
column1=int(input("Enter the value of column of matrix1:"))
matrix2=[]
row2=int(input("Enter the value of row of matrix2:"))
column2=int(input("Enter the value of column of matrix2:"))
if(row1!=row2 or column1!=column2):
  print("Wrong input")
else:
  #Input for matrix 1
  for a in range(row1):
    empty=[]
    for b in range(column1):
      x=int(input("Enter the element in matrix 1:"))
      empty.append(x)
    matrix1.append(empty)

  #Input for matrix 2
  for a in range(row2):
    empty=[]
    for b in range(column2):
      x=int(input("Enter the element in matrix 2:"))
      empty.append(x)
    matrix2.append(empty)

  # Addition of two matrix
  matrix3=[]
  for a in range(row1):
    empty=[]
    for b in range(column2):
      x=matrix1[a][b]+matrix2[a][b]
      empty.append(x)
    matrix3.append(empty)
  print(matrix3)
```

Calculation of mean squared error and mean valued error

```python
#Question 8
list1=[]
len1=int(input("Enter the length of original list:"))
for a in range(len1):
  x=int(input("Enter the element in list 1"))
  list1.append(x)
list2=[]
len2=int(input("Enter the length of predicted list:"))
for a in range(len2):
  x=int(input("Enter the element of predicted list:"))
  list2.append(x)
if(len1!=len2):
  print("Wrong input")
else:
  # Calculation of Mean Absolute Error
  mae=0
  for a in range(len1):
    mae+=abs(list1[a]-list2[a])
  mae=mae/len1
  # Calculation of Mean Squared Error
  mse=0
  for a in range(len2):
    mse+=(list1[a]-list2[a])**2
  mse=mse/len2
  print(mae)
  print(mse)
```
